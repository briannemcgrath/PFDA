```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = sns.load_dataset('tips')
dataset.head()
```

```python
sns.set_style('whitegrid')
sns.lmplot(x='total_bill', y='tip', order=1, data=dataset, line_kws={'color': 'grey', 'linewidth':2},
           scatter_kws={'color': 'lightblue', 's': 50, 'alpha': 0.7})

plt.title('Total Bill v Tip', fontsize=12, weight='bold')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
sns.lmplot(x='size', y='tip', data=dataset, x_estimator=np.mean,line_kws={'color': 'lightblue', 'linewidth':2},
           scatter_kws={'color': 'grey', 's': 50, 'alpha': 0.7})
```
